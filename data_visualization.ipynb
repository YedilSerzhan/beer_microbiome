```python
class DataSample:
  def __init__(self, sample_num, sample_type, sample_location):
    self.sample_num = sample_num
    self.sample_type = sample_type
    self.sample_location = sample_location
  def __str__(self):
    return f"{self.sample_type}, {self.sample_num}, {self.sample_location}"

class Dataset:
  def __init__(self, title, date, sequencing_types, sequencing_technique, targets, samples):
    self.title = title
    self.date = date
    self.sequencing_types = sequencing_types
    self.sequencing_technique = sequencing_technique
    self.targets = targets
    self.samples = samples
  def __str__(self):
    return f"{self.title}, {self.date}, {self.samples}"
```

```python
all_sample_types = []
sample_types_to_num = {}
sequencing_types_to_num = {}
targets_to_num = {}
for d in data:
    this_sample_num = 0
    for sample in d.samples:
        this_sample_num += sample.sample_num
        if sample.sample_type not in sample_types_to_num:
            sample_types_to_num[sample.sample_type] = 0
        sample_types_to_num[sample.sample_type] += sample.sample_num
    for sequencing_type in d.sequencing_types:
        if sequencing_type not in sequencing_types_to_num:
            sequencing_types_to_num[sequencing_type] = 0
        sequencing_types_to_num[sequencing_type] += this_sample_num
    for target in d.targets.values():
        if target not in targets_to_num:
            targets_to_num[target] = 0
        targets_to_num[target] += this_sample_num        

print(sample_types_to_num)
print("There are", sum(list(sample_types_to_num.values())), "samples in total.")
print(sequencing_types_to_num)
print(targets_to_num)

import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy as np

plt.style.use('_mpl-gallery')

plt.figure(0)
# data
x = list(range(1, 1+len(sample_types_to_num)))
y = list(sample_types_to_num.values())
# plot
fig, ax = plt.subplots(figsize=(4, 4))

ax.bar(x, y)
plt.xticks(x, list(sample_types_to_num.keys()))
plt.xticks(rotation=90, fontsize = 'small')
plt.show()
```

```python
plt.figure(2)
# Creating dataset
targets = list(targets_to_num.keys())
 
target_counts = list(targets_to_num.values())
 
 
# Creating explode data
explode = (0.0, 0.0, 0.0, 0.2, 0.0)
 
# Creating color parameters
colors = ( "orange", "cyan", "brown",
          "grey", "indigo")
 
# Wedge properties
wp = { 'linewidth' : 1, 'edgecolor' : "green" }
 
# Creating autocpt arguments
def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:.1f}%\n({:d} g)".format(pct, absolute)
 
# Creating plot
fig, ax = plt.subplots(figsize =(10, 7))
wedges, texts, autotexts = ax.pie(target_counts,
                                  autopct = lambda pct: func(pct, target_counts),
                                  explode = explode,
                                  labels = targets,
                                  shadow = True,
                                  colors = colors,
                                  startangle = 90,
                                  wedgeprops = wp,
                                  textprops = dict(color ="magenta"))
 
# Adding legend
ax.legend(wedges, targets,
          title ="Targets",
          loc ="center left",
          bbox_to_anchor =(1, 0, 0.5, 1))
 
plt.setp(autotexts, size = 8, weight ="bold")
ax.set_title("Amplicon targets")
 
# show plot
plt.show()
```
